```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("amazon.csv")
df.head()
```

```python
df.info()
```

```python
df.describe(include='all')
```

```python
df['discounted_price'] = (
    df['discounted_price']
    .str.replace('[^0-9.]', '', regex=True)
)

df['actual_price'] = (
    df['actual_price']
    .str.replace('[^0-9.]', '', regex=True)
)
```

```python
def corrigir_preco(x):
    if x.count('.') > 1:
        partes = x.split('.')
        return partes[0] + ''.join(partes[1:-1]) + '.' + partes[-1]
    return x
```

```python
df['discounted_price'] = df['discounted_price'].apply(corrigir_preco)
df['actual_price'] = df['actual_price'].apply(corrigir_preco)
```

```python
df['discounted_price'] = df['discounted_price'].astype(float)
df['actual_price'] = df['actual_price'].astype(float)
```

```python
df['discount_percentage'] = (
    df['discount_percentage']
    .str.replace('[^0-9.]', '', regex=True)
    .astype(float)
)
```

```python
df['rating'] = pd.to_numeric(df['rating'], errors='coerce')
df['rating_count'] = pd.to_numeric(df['rating_count'], errors='coerce')
```

```python
df = df.dropna()
```

```python
df[['discounted_price', 'actual_price', 'rating']].head()
df.describe()
```

```python
df = df.rename(columns={
    'product_id': 'id_produto',
    'product_name': 'nome_produto',
    'category': 'categoria',
    'discounted_price': 'preco_com_desconto',
    'actual_price': 'preco_original',
    'discount_percentage': 'percentual_desconto',
    'rating': 'avaliacao_media',
    'rating_count': 'quantidade_avaliacoes',
    'about_product': 'descricao_produto',
    'user_id': 'id_usuario',
    'user_name': 'nome_usuario',
    'review_id': 'id_avaliacao',
    'review_title': 'titulo_avaliacao',
    'review_content': 'conteudo_avaliacao',
    'img_link': 'link_imagem',
    'product_link': 'link_produto'
})
```

```python
df.columns = [c.lower() for c in df.columns]
```

```python
df.columns
```

```python
df.columns = df.columns.str.lower().str.strip()
df.columns
```

```python
df['preco_com_desconto'] = df['preco_com_desconto'].astype(str)
df['preco_com_desconto'] = df['preco_com_desconto'].str.replace('.', '', regex=False)
df['preco_com_desconto'] = df['preco_com_desconto'].str.replace(',', '.', regex=False)
df['preco_com_desconto'] = pd.to_numeric(df['preco_com_desconto'], errors='coerce')
```

```python
df['avaliacao_media'] = pd.to_numeric(df['avaliacao_media'], errors='coerce')
```

```python
df = df.dropna(subset=['preco_com_desconto', 'avaliacao_media'])
```

```python
plt.figure()
plt.hist(df['preco_com_desconto'], bins=30)
plt.title('Distribuição de Preços dos Produtos')
plt.xlabel('Preço com Desconto (R$)')
plt.ylabel('Quantidade de Produtos')
plt.show()
```

```python
plt.figure()
plt.hist(df['avaliacao_media'], bins=10)
plt.title('Distribuição das Avaliações dos Produtos')
plt.xlabel('Avaliação Média (Estrelas)')
plt.ylabel('Quantidade de Produtos')
plt.show()
```

```python
plt.figure()
plt.scatter(df['preco_com_desconto'], df['avaliacao_media'])
plt.title('Relação entre Preço e Avaliação dos Produtos')
plt.xlabel('Preço com Desconto (R$)')
plt.ylabel('Avaliação Média (Estrelas)')
plt.show()
```

```python
top_avaliados = (
    df.groupby('nome_produto')['quantidade_avaliacoes']
    .sum()
    .sort_values(ascending=False)
    .head(10)
)

plt.figure()
plt.barh(top_avaliados.index, top_avaliados.values)
plt.title('Top 10 Produtos com Maior Número de Avaliações')
plt.xlabel('Quantidade de Avaliações')
plt.ylabel('Produto')
plt.show()
```
